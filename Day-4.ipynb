```python
events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True)
```

```python
events.write.format("delta").mode("overwrite").save("/Volumes/workspace/ecommerce/ecommerce_data/delta/ecommerce_events")
```

```python
events.write.format("delta").saveAsTable("events_table")
```

```python
spark.sql("""
    CREATE OR REPLACE TABLE events_delta
    USING DELTA
    AS SELECT * FROM events_table
""")

```

```python
try:
    wrong_schema = spark.createDataFrame(
        [("a", "b", "c")],
        ["x", "y", "z"]
    )

    wrong_schema.write.format("delta") \
    .mode("append") \
    .save("/Volumes/workspace/ecommerce/ecommerce_data/delta/ecommerce_events")

except Exception as e:
    print("Schema enforcement working")
    print(e)
```
