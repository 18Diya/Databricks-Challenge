```python
# Explain query
spark.sql("SELECT * FROM default.silver_events WHERE event_type='purchase'").explain(True)
```

```python
spark.sql("""
CREATE TABLE workspace.default.events_part
USING DELTA
PARTITIONED BY (event_date, event_type)
AS
SELECT * FROM workspace.default.silver_events
""")
```

```python
# Optimize
spark.sql("OPTIMIZE default.events_part ZORDER BY (user_id, product_id)")
```
